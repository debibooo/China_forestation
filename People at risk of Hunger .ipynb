```python
import numpy as np
from scipy.stats import norm
import pandas as pd

# Correct the file path (removed the extra quote at the end)
excel_path = "/.../cal_11%.xlsx"

# Read the Excel file, loading all sheets into a dictionary of DataFrames
sheets_dict = pd.read_excel(excel_path, sheet_name=None)


with pd.ExcelWriter('/Volumes/UCL/论文工作/forestation/my_output_folder/PoU_calculations_11%.xlsx') as writer:
    # 遍历所有sheets
    for sheet_name, df in sheets_dict.items():
        
        # 计算PoU
        df["sigma_t"] = np.sqrt(np.log(df['CV']**2 + 1))
       
        df['mu_calt_sigma_t'] = np.log(df['cal']) - (df["sigma_t"]**2 / 2)

        df['PoU'] = norm.cdf((np.log(df['Mt']) - df['mu_calt_sigma_t']) / df["sigma_t"])
        df['risk'] =  df['PoU'] * df['POP'] 

        # save the hunger risk results to the same sheet in the new Excel file
        df.to_excel(writer, sheet_name=sheet_name)
```

```python

import numpy as np
import scipy.stats as stats

def calculate_pou(cal_t, M_t, CV):
    # Step 1: Calculate sigma_t using equation (4)
    sigma_t = np.sqrt(np.log(CV**2 + 1))

    # Step 2: Calculate mu(cal_t, sigma_t) using equation (3)
    mu_t = np.log(cal_t) - (sigma_t**2 / 2)

    # Step 3: Calculate PoU_t using equation (2)
    z_score = np.log(M_t) - (mu_t / sigma_t)
    pou_t = stats.norm.cdf(z_score)  # Cumulative distribution function of standard normal

    return pou_t

# Example usage
cal_t = 2500  # Average dietary energy supply (in kcal/person/day)
M_t = 2100    # Mean minimum dietary energy requirement (in kcal/person/day)
CV = 0.2      # Coefficient of variation

pou_t = calculate_pou(cal_t, M_t, CV)
print(f"Probability of Undernourishment (PoU) in year t: {pou_t:.4f}")
```
